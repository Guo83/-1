```python
import numpy as np
from numpy.linalg import inv
import matplotlib.pyplot as plt

def LinearRegression(x,y):
    x = x.T; y = y.T
    weight = np.dot(np.dot(inv(np.dot(x.T, x)), x.T), y)
    y_predict = np.dot(x,weight)
    return (y_predict,weight)
x = np.array([[4.2,7.1,6.3,1.1,0.2,4.0,3.5,8,2.3],[1] * 9])
y = np.array([[8600,6100,6700,12000,14200,8500,8900,6200,11200]])

plt.xlabel("D/(km)")
plt.ylabel("R/(元/m^2)")
plt.plot(x[0],LinearRegression(x,y)[0],color = 'k')
plt.plot(x[0],y[0],'o')
plt.show()
```

```python
import random

def testLiner(range_down,range_up,step,noise):
    x = np.empty([step,1],dtype=float)
    y = np.empty([step,1],dtype=float)
    for i in range(step):
        noise1 = random.uniform(-noise,noise)
        x[i][0] = random.uniform(range_down,range_up)
        y[i][0] = 3*x[i][0]+10+noise1
    return x,y


x,y = testLiner(-10,10,1000,10)
plt.scatter(x,y)
test_0 = LinearRegression(x,y)
```
